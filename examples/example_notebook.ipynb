# Exemples d'utilisation de casabourse
Rappels : exécuter les cellules une par une.

```python
import casabourse as cb
from IPython.display import display
print('build id:', cb.get_build_id_cached())
```

```python
# Live market
df = cb.get_live_market_data()
display(df.head() if df is not None else 'no data')
```

```python
# Données historiques pour un ticker
hist = cb.get_historical_data('IAM', '2024-01-01', '2024-12-31')
display(hist.head() if hist is not None else 'no data')
```

```python
# Volumes et capitalisation
gv = cb.get_volume_overview()
if gv and gv[0] is not None:
    display(gv[0])
cap = cb.get_capitalization_overview()
if cap:
    display(cap['global_cap'])
```

```python
# Indices : composition et cotation
comp = cb.get_index_composition('MSI20')
display(comp.head() if comp is not None else 'no composition')
quot = cb.get_index_quotation('MSI20')
display(quot if quot is not None else 'no quotation')
```
